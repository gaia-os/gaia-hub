
```
    ./data/projects/Google/AmazonReforestation.json
```

````python
# Import required classes from the natural_models package
from natural_models.agents.AgentFactory import AgentFactory
from natural_models.assessment.DataLoader import DataLoader
from natural_models.kernels.impl.MCMCKernel import MCMCKernel
from natural_models.kernels.impl.SVIKernel import SVIKernel
from natural_models.visualisation.distributions import draw_beliefs
from natural_models.visualisation.distributions import compare_posteriors
````

### Loading data
The class responsible for loading the data from the filesystem is called `DataLoader`.

It requires the path to the project's configuration file, and provides access to
* lot info
* available reports
* policy to evaluate
* more

````python
# Load the data corresponding to the ElementSix model
data_loader = DataLoader('../data/projects/Roots-and-Culture/roots-indoor1.json')
````

### Creating compatible agents
The class responsible for instantiating the agents compatible with the project configuration file is called: `AgentFactory`.

All compatible agents can be retrieved by calling the method AgentFactory.create, and providing the data loader as parameter.

Specific agents can be recovered using their unique names, e.g., the deterministic model of Element6 is called:
```
    Element6.WisconsinFarm.Deterministic
```

````python
# Load the agents agent(s) compatible with the loaded data
agents = AgentFactory.create(data_loader, verbose=True)

# Get ordinary differential equations (ODE) and stochastic differential equations (SDE) agents
ode_agent = next(filter(lambda a: a.name == "Roots-and-Culture.roots-indoor1.Deterministic", agents))
sde_agent = next(filter(lambda a: a.name == "Roots-and-Culture.roots-indoor1.Stochastic", agents))
````

### Performing predictions
The agent class provides a method called `predict` that calls the `Predictive` numpyro class, and performs prediction into the future based on the policy described in the project configuration.

The `predict` function takes as paramters the agent's model that will be used to perform prediction into the future. This function returns samples from the distribution over future random variables, and the number of samples returned for each variable can be specified as parameter.

````python
# Predict the future using the ODE and SDE agents
ode_prediction_samples = ode_agent.predict(model=ode_agent.model, num_samples=1)
sde_prediction_samples = sde_agent.predict(model=sde_agent.model, num_samples=1)
````

### Visualize the predictions

````python
# Draw prior beliefs using predictive samples
fig = draw_beliefs(ode_prediction_samples)
````

### Performing inference

First, let's provide the agent with the data and policy that it should use to perform inference.

````python
# Provide the policy to the deterministic agent
ode_agent.policy = ode_agent.policy[:, -1:]

# Provide the data to the deterministic agent
data = {
    'lai': ode_prediction_samples['lai'][0]
}
cond_model, cond_guide = ode_agent.condition_all(data=data)
````

Second, let's create the mask specifying which observations are provided to the model:

````python
# The measurement interval is the number of time steps between two observations made by the agent 
measurement_interval = 2

# Create the mask
mask = {
    'lai': jnp.expand_dims(jnp.arange(len(data['lai'])), -1) % measurement_interval == 0
}
````

Third, let's run inference using Monte Carlo Markov Chain (MCMC):

````python
# Create the MCMC algorithm
mcmc_args = {
    "num_warmup": 1000,
    "num_samples": 1000, 
    "num_chains": 1
}
mcmc_algorithm = ode_agent.inference_algorithm(MCMCKernel, kernel_args=mcmc_args)

# Perform inference using MCMC
inference_args = {
    "mask": mask,
    "data_level": 0
}
mcmc_samples = mcmc_algorithm.run_inference(inference_args=inference_args)
````

Fourth, let's run inference using Stochastic Variational Inference (SVI):

````python
# Create the SVI algorithm
svi_args = {
    "optimizer": optax_to_numpyro(optax.adabelief(1e-3, eps=1e-8, eps_root=1e-8)),
    "loss": Trace_ELBO(num_particles=10)
}
svi_algorithm = ode_agent.inference_algorithm(SVIKernel, kernel_args=svi_args)

# Perform inference using SVI
inference_params = {
    "stable_update": True, 
    "mask": mask,
    "data_level": 0
}
prediction_params = {
    "num_step": 20000
}
svi_samples = svi_algorithm.run_inference(inference_params=inference_params, prediction_param=prediction_params)
````

### Visualize the posterior beliefs

Call the function in charge of comparing the initial predictions and the posterior distribution, i.e. MCMC and SVI posteriors.

````python
# Compare the initial prediction, the SVI posterior, and the MCMC posterior
compare_posteriors(
    is_observed=mask['lai'],
    mcmc_samples=mcmc_samples,
    svi_samples=svi_samples,
    prediction_samples=ode_prediction_samples
)
````
